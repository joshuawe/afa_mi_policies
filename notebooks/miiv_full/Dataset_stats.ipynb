```python
import numpy as np
import pandas as pd
```

```python
path = '../../data/miiv_full/MCAR_1/miiv_MCAR_1_ts_wide.parquet'

df = pd.read_parquet(path)
```

```python
grouped = df.groupby('id')
lengths = grouped.size()

ax = lengths.hist(bins=100)
ax.set_title('Length of Patient stays')
ax.set(xlabel='Length of stay [days]', ylabel='Num patients (histogram)', xlim=(0,170))

# add cumulative distribution
ax2 = ax.twinx()  # Create a twin y-axis sharing the x-axis
ax2.plot(np.sort(lengths), np.linspace(0, 1, len(lengths), endpoint=False), color='red')
ax2.set_ylabel('CDF (line)')
ax2.set_ylim([0, 1])
```

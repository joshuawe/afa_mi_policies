# Visualizing the performance plots of AFA agents

Focusing on the performance of the predictor and visualizing the plots next to each other

```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def plot_images(root_folder, image_names):
    """
    Plots PNG images from the given root folder and subfolders for the specified image names.
    
    Parameters:
    - root_folder (str): The root folder where to start the search
    - image_names (list): List of image names to search for and plot
    """
    for image_name in image_names:
        # Initializing a list to store image paths
        image_paths = []
        
        # Traverse the directory structure to find PNG images
        for root, dirs, files in os.walk(root_folder):
            for file in files:
                if file == image_name:
                    image_path = os.path.join(root, file)
                    image_paths.append(image_path)
                    
        # Plot the images side by side
        plt.figure(figsize=(15, 5))
        for i, img_path in enumerate(image_paths):
            img = mpimg.imread(img_path)
            plt.subplot(1, len(image_paths), i + 1)
            plt.imshow(img)
            plt.axis('off')
            plt.title(f"{image_name} from {os.path.dirname(img_path).split('/')[-1]}")
        plt.show()
```

```python
def plot_images_by_paths(root_folder, image_names, paths):
    """
    Plots PNG images from the given root folder, subfolders, and specified paths for the provided image names.
    Each row of subplots corresponds to images from a specific path.
    
    Parameters:
    - root_folder (str): The root folder where to start the search
    - image_names (list): List of image names to search for and plot
    - paths (list): List of specific paths to look for images
    """
    
    for path in paths:
        max_width = 0  # Initialize max width to adjust figure size
        
        for image_name in image_names:
            # Initialize a list to store image paths
            image_paths = []
            
            # Traverse the directory structure to find PNG images
            for root, dirs, files in os.walk(os.path.join(root_folder, path)):
                for file in files:
                    if file == image_name:
                        image_path = os.path.join(root, file)
                        image_paths.append(image_path)
            
            # Update max width
            max_width = max(max_width, len(image_paths))
            
            # Plot the images side by side for each path
        plt.figure(figsize=(max_width * 5, 5))
        plt.suptitle(f"Images from {path}")
        
        for i, image_name in enumerate(image_names):
            image_paths = []
            
            # Traverse the directory structure to find PNG images
            for root, dirs, files in os.walk(os.path.join(root_folder, path)):
                for file in files:
                    if file == image_name:
                        image_path = os.path.join(root, file)
                        image_paths.append(image_path)
            
            for j, img_path in enumerate(image_paths):
                img = mpimg.imread(img_path)
                plt.subplot(len(image_names), max_width, i * max_width + j + 1)
                plt.imshow(img)
                plt.axis('off')
                plt.title(f"{image_name} from {os.path.dirname(img_path).split('/')[-1]}")
        plt.show()

# Sample usage:
# plot_images_by_paths("root", ["plotA.png", "plotB.png"], ["model1", "model2"])
```

```python
def plot_images_by_root_paths(root_folders, image_names):
    """
    Plots PNG images from the given root folders and subfolders for the specified image names.
    Each row of subplots corresponds to images from a specific root folder.
    
    Parameters:
    - root_folders (list): List of root folders where to start the search
    - image_names (list): List of image names to search for and plot
    """
    
    for root_folder in root_folders:
        max_width = 0  # Initialize max width to adjust figure size
        
        # Loop over image names to find the maximum width for subplots
        for image_name in image_names:
            # Initialize a list to store image paths
            image_paths = []
            
            # Traverse the directory structure to find PNG images
            for root, dirs, files in os.walk(root_folder):
                for file in files:
                    if file == image_name:
                        image_path = os.path.join(root, file)
                        image_paths.append(image_path)
            
            # Update max width
            max_width = max(max_width, len(image_paths))
        
        # Plot the images side by side for each root folder
        plt.figure(figsize=(max_width * 5, 5))
        plt.suptitle(f"Images from {root_folder}")
        
        for i, image_name in enumerate(image_names):
            image_paths = []
            
            # Traverse the directory structure to find PNG images
            for root, dirs, files in os.walk(root_folder):
                for file in files:
                    if file == image_name:
                        image_path = os.path.join(root, file)
                        image_paths.append(image_path)
            
            for j, img_path in enumerate(image_paths):
                img = mpimg.imread(img_path)
                plt.subplot(len(image_names), max_width, i * max_width + j + 1)
                plt.imshow(img)
                plt.axis('off')
                plt.title(f"{image_name} from {os.path.dirname(img_path).split('/')[-1]}")
        plt.show()

# Sample usage:
# plot_images_by_root_paths(["root1", "root2"], ["plotA.png", "plotB.png"])
```

```python
def plot_images_by_root_paths_single_figure(root_folders, image_names):
    """
    Plots PNG images from the given root folders and subfolders for the specified image names.
    All images are plotted in a single figure.
    
    Parameters:
    - root_folders (list): List of root folders where to start the search
    - image_names (list): List of image names to search for and plot
    """
    
    global_max_width = 0  # Initialize global max width to adjust figure size
    
    # Loop over root folders and image names to find the global maximum width for subplots
    for root_folder in root_folders:
        for image_name in image_names:
            # Initialize a list to store image paths
            image_paths = []
            
            # Traverse the directory structure to find PNG images
            for root, dirs, files in os.walk(root_folder):
                for file in files:
                    if file == image_name:
                        image_path = os.path.join(root, file)
                        image_paths.append(image_path)
            
            # Update global max width
            global_max_width = max(global_max_width, len(image_paths))
    
    # Create a single figure for all root folders
    plt.figure(figsize=(global_max_width * 5, len(root_folders) * len(image_names) * 5))
    
    # Plot images
    subplot_index = 1
    for i, root_folder in enumerate(root_folders):
        for j, image_name in enumerate(image_names):
            image_paths = []
            
            # Traverse the directory structure to find PNG images
            for root, dirs, files in os.walk(root_folder):
                for file in files:
                    if file == image_name:
                        image_path = os.path.join(root, file)
                        image_paths.append(image_path)
            
            for k, img_path in enumerate(image_paths):
                img = mpimg.imread(img_path)
                plt.subplot(len(root_folders) * len(image_names), global_max_width, subplot_index)
                plt.imshow(img)
                plt.axis('off')
                
                # Add title only for the first subplot in each row
                if k == 0:
                    plt.title(f"{image_name} from {os.path.dirname(img_path).split('/')[-1]} in {root_folder[-21:]}")
                subplot_index += 1
    plt.show()

# Sample usage:
# plot_images_by_root_paths_single_figure(["root1", "root2"], ["plotA.png", "plotB.png"])
```

```python
def plot_images_by_root_paths_single_figure(root_folders, image_names):
    """
    Plots PNG images from the given root folders and subfolders for the specified image names.
    All images are plotted in a single figure.
    
    Parameters:
    - root_folders (list): List of root folders where to start the search
    - image_names (list): List of image names to search for and plot
    """
    
    global_max_width = 0  # Initialize global max width to adjust figure size
    
    # Loop over root folders and image names to find the global maximum width for subplots
    for root_folder in root_folders:
        for image_name in image_names:
            # Initialize a list to store image paths
            image_paths = []
            
            # Traverse the directory structure to find PNG images
            for root, dirs, files in os.walk(root_folder):
                for file in files:
                    if file == image_name:
                        image_path = os.path.join(root, file)
                        image_paths.append(image_path)
            
            # Update global max width
            global_max_width = max(global_max_width, len(image_paths))
    
    # Create a single figure for all root folders
    fig, axs = plt.subplots(len(root_folders) * len(image_names), global_max_width, figsize=(global_max_width * 5, len(root_folders) * len(image_names) * 5))
    
    # Plot images
    subplot_index = 0
    for i, root_folder in enumerate(root_folders):
        for j, image_name in enumerate(image_names):
            image_paths = []
            
            # Traverse the directory structure to find PNG images
            for root, dirs, files in os.walk(root_folder):
                for file in files:
                    if file == image_name:
                        image_path = os.path.join(root, file)
                        image_paths.append(image_path)
            
            for k, img_path in enumerate(image_paths):
                img = mpimg.imread(img_path)
                ax = axs[subplot_index, k]
                ax.imshow(img)
                ax.axis('off')
                
                # Add title to each subplot
                ax.set_title(f"{os.path.dirname(img_path).split('/')[-1]}")
                
                # Add title to the first subplot in each row
                if k == 0:
                    ax.set_title(f"{image_name} in {root_folder[-22:]}\n{os.path.dirname(img_path).split('/')[-1]}")
            subplot_index += 1
    plt.tight_layout()
    plt.show()

# Sample usage:
# plot_images_by_root_paths_single_figure(["root1", "root2"], ["plotA.png", "plotB.png"])
```
